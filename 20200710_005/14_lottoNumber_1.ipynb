```python
import random # 난수를 발생시키는 randrange() 메소드를 사용하려면 random 모듈을 import 시켜야 한다.
import time   # 프로그램을 일시적으로 멈추는 sleep() 메소드를 사용하려면 time 모듈을 import 시켜야 한다.
```

```python
# 로또 추첨기로 사용할 i가 45번 반복되는 리스트를 만든다.
lotto = [i for i in range(1, 46)]
# print(lotto)

'''
# 섞기 전 상태를 출력한다. => 1줄에 10개씩 2자리로 출력한다.
for i in range(len(lotto)):
    print("%2d " % lotto[i], end = "")
    if (i + 1) % 10 == 0:
        print()
    # ===== if (i + 1) % 10 == 0: 끝 =====
# ===== for i in range(len(lotto)): 끝 =====
print()
'''

# 섞는다.
# lotto[0]와(고정) lotto[1] ~ lotto[44] 중에서 랜덤한 위치를 선택해서 두 기억장소의 데이터를 교환한다.
for i in range(1000000):
    # lotto[1] ~ lotto[44] 중에서 랜덤한 위치를 선택한다.
    # randrange(a, b) => a 부터 b-1 사이의 숫자를 무작위로 추출한다.
    r = random.randrange(1, 45)
    # lotto[0]와 lotto[r]에 저장된 데이터를 교환한다.
    # temp = lotto[0]
    # lotto[0] = lotto[r]
    # lotto[r] = temp
    lotto[0], lotto[r] = lotto[r], lotto[0]
# ===== for i in range(1000000): =====

'''
# 섞은 후 상태를 출력한다. => 1줄에 10개씩 2자리로 출력한다.
for i in range(len(lotto)):
    print("%2d " % lotto[i], end = "")
    if (i + 1) % 10 == 0:
        print()
    # ===== if (i + 1) % 10 == 0: 끝 =====
# ===== for i in range(len(lotto)): 끝 =====
print()
'''

# 1등 번호와 보너스 번호를 출력한다.
print("1등 번호 : ", end = "")
for i in range(6):
    print("{0:2d} ".format(lotto[i]), end = "")
    # time 모듈의 sleep() 메소드는 인수로 지정된 시간만큼 프로그램을 일시적으로 멈춘다.
    # sleep() 메소드에 프로그램을 멈추는 시간은 "초" 단위로 지정한다.
    time.sleep(1)
# ===== for i in range(6): 끝 =====
print("보너스 : {0:2d}".format(lotto[6]))
```
