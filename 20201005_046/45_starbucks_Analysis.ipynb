```python
import warnings
warnings.filterwarnings('ignore')
# 경고 메시지를 표시하지 않으려면 위의 2줄을 코딩한다. => 다시 경고 메시지를 보고싶다면 warnings.filterwarnings('default')를 실행하면 출력된다.
import requests
import folium
import json
from pandas.io.json import json_normalize
```

```python
# 전국의 특정 지역 스타벅스 매장을 찾아서 지도 위에 표시하기
# 1. requests로 스타벅스 매장 위치 데이터 가져오기
# 2. json_normalize를 통해서 json 형태의 데이터를 pandas 데이터프레임으로 변환한다.
# 3. folium을 사용해서 지도 위에 마커를 표시한다.
```

```python
targetSite = 'https://www.starbucks.co.kr/store/getStore.do?r=IVIVA0L5E4'
req = requests.post(targetSite, data = {
    'p_sido_cd': '01',
    'p_gugun_cd': '',
    'ins_lat': 37.5108295,
    'ins_lng': 127.0292880,
    'in_biz_cd': '',
    'set_date': '',
    'iend': 1500
})

print(type(req.text))
print(req.text)
```

```python
# json 모듈의 loads() 메소드로 스타벅스 매장 위치 정보가 저장된 json 타입의 데이터를 파이썬에서 처리하기 위해 딕셔너리 타입으로 변환한다.
star_json = json.loads(req.text)
print(type(star_json))
print(star_json)
```

```python
# pandas.io.json 모듈의 json_normalize() 메소드로 json 타입의 데이터를 변환한 딕셔너리를 판다스 데이터프레임으로 변환한다.
star_df = json_normalize(star_json, 'list')
print(type(star_df))
print(len(star_df))    # 행(row)의 개수, 서울에 2020년 10월 5일 현재 524개의 매장이 있다.
print(star_df.columns) # 열(column)의 개수는 122개
```

```python
# 주요한 column 몇 가지를 선택해서 지도에 marker를 표시할 때 사용할 데이터가 저장된 데이터프레임을 만든다.
# s_name       => 지점 이름
# sido_code    => 시도 코드
# sido_name    => 시도 이름
# gugun_code   => 구군 코드
# gugun_name   => 구군 이름
# doro_address => 도로명 주소
# lat          => 위도
# lot          => 경도
star_df_map = star_df[['s_name', 'lat', 'lot', 'sido_name', 'gugun_name', 'doro_address']]
star_df_map
```

```python
# 위도, 경도 데이터 타입을 object(판다스에서 문자열을 의미한다.)에서 float(실수)로 변환한다.
star_df_map['lat'] = star_df_map['lat'].astype(float)
star_df_map['lot'] = star_df_map['lot'].astype(float)
star_df_map.dtypes
```

```python
# 종각점
star_df_map[star_df_map['s_name'] == '종각']
```

```python
# 광교점
star_df_map[star_df_map['s_name'] == '광교']
```
