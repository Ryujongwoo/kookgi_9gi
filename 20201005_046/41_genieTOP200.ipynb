```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
import time
```

```python
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20201005&hh=09&rtm=Y&pg=' # 공통 부분
for i in range(1, 5):
    print(url + str(i))
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20201005&hh=09&rtm=Y&pg=1 =>   1 ~  50위 url
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20201005&hh=09&rtm=Y&pg=2 =>  51 ~ 100위 url
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20201005&hh=09&rtm=Y&pg=3 => 101 ~ 150위 url
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20201005&hh=09&rtm=Y&pg=4 => 151 ~ 200위 url
header = {"User-agent": "Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko"}
```

```python
# 노래 제목 크롤링
for i in range(1, 5):
    targetSite = '{}{}'.format(url, i)
    # print(targerSite)
    req = requests.get(targetSite, headers = header)
    # print(req)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    # print(soup)
    time.sleep(1)
    titles = soup.findAll('a', {'class': 'title ellipsis'})
    # print(len(titles))
    for title in titles:
        # print(title.text.strip())
        # print(title.text.split('\n')[-1].strip())
        print(title.text.strip().split('\n')[0])
```

```python
# 아티스트 크롤링
for i in range(1, 5):
    targetSite = '{}{}'.format(url, i)
    req = requests.get(targetSite, headers = header)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    time.sleep(1)
    artists = soup.findAll('a', {'class': 'artist ellipsis'})
    # print(len(artists))
    for artist in artists[5:]:
        # print(artist.text)
        print(artist.text.strip().split('\n')[0])
    print('=' * 80)
```

```python
genie = []
for i in range(1, 5):
    targetSite = '{}{}'.format(url, i)
    req = requests.get(targetSite, headers = header)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    time.sleep(1)
    titles = soup.findAll('a', {'class': 'title ellipsis'})
    artists = soup.findAll('a', {'class': 'artist ellipsis'})
    for i in range(len(titles)):
        genie.append([titles[i].text.strip().split('\n')[0], artists[i + 5].text.strip().split('\n')[0]])
print(genie)
```

```python
print(dt.now(), '현재')
for i in range(len(genie)):
    print('{0:3d}위. {1} - {2}'.format(i + 1, genie[i][0], genie[i][1]))
```

```python
file = open('genieTOP200.txt', 'w')
file.write('{} 현재 Genie 뮤직 실시간 TOP 200\n'.format(dt.now()))
for i in range(len(genie)):
    data = '{0:3d}위 {1} - {2}'.format(i + 1, genie[i][0], genie[i][1])
    file.write(data + '\n')
file.close()
print('genieTOP200.txt 파일에 쓰기 완료')
```
