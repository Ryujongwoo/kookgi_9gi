```python
# 모든 프로그래밍 언어는 함수의 실행 결과를 return 뒤에 1개만 붙여서 return 시킬 수 있다. => 파이썬도 예외는 아니다.
def swap(a, b):
    temp = a
    a = b
    b = temp
    # 데이터를 ","로 구분해서 나열하면 양쪽 끝에 ()를 입력하지 않아도 튜플로 만들어진다.
    # return을 튜플로 하면 return 값이 2개 이상인 것 같은 효과를 낸다. => 공식적으로 튜플 1개를 리턴한다.
    return a, b # (a, b) 형태의 튜플 1개를 return 시킨다. => 마치 return 값이 2개인것 처럼 보인다.
```

```python
a = 3; b = 4
print("a = {}, b = {}".format(a, b))
a, b = b, a
print("a = {}, b = {}".format(a, b))
a, b = swap(a, b)
print("a = {}, b = {}".format(a, b))
result = swap(a, b)
print(type(result))
print("a = {}, b = {}".format(result[0], result[1]))
```
