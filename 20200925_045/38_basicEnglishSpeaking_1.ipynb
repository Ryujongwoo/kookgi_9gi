```python
import requests
from bs4 import BeautifulSoup
```

```python
targerSite = 'https://basicenglishspeaking.com/daily-english-conversation-topics/'
req = requests.get(targerSite)
# print(req)
html = req.text
# print(html)
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
# 파싱한 대화 주제를 저장할 빈 리스트를 만든다.
subjects = []
divs = soup.findAll('div', {'class': 'tcb-col'})
# print(len(divs))
for div in divs:
    # print(len(div))
    chapters = div.findAll('a')
    # print(len(chapters))
    for i in range(len(chapters)):
        subjects.append(chapters[i].text)
```

```python
for i in range(len(subjects)):
    print('{0:2d}. {1}'.format(i + 1, subjects[i]))
```
