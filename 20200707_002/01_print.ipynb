```python
# print() 함수는 () 안의 내용을 출력하고 줄을 바꾼다.
print("Hello Python")
print('안녕 파이썬')
print("제2외국어를 배운 호랑이가 '야옹' 했습니다.")
print('제2외국어를 배운 고양이가 "어흥" 했습니다.')
print("제2외국어를 배운 호랑이가 \"야옹\" 했습니다.")
print('제2외국어를 배운 고양이가 \'어흥\' 했습니다.')
```

```python
# print() 함수로 출력 후 줄을 변경하지 않으려면 아래와 같이 end 옵션을 지정한다.
print("Hello \nPython", end = ", ") # \n => new line => 줄을 바꾼다.
print('안녕 파이썬')
# end 옵션은 print() 함수로 출력한 후 출력할 내용을 지정한다. => end 옵션을 생략하면 "\n"이 기본값으로 사용된다.
```

```python
# print() 함수로 2개 이상의 데이터를 출력하려면 ","나 "+"로 구분해서 출력한다.
print("Hello Python", "안녕 파이썬") # ","로 구분하면 데이터 사이에 공백이 출력된다.
print("Hello Python" + "안녕 파이썬") # "+"로 구분하면 데이터가 연결되서 출력된다.
# print() 함수에서 문자열 뒤에 "*" 연산자를 사용하면 지정된 숫자만큼 문자열을 반복해서 출력한다.
print("^^;" * 50)
```

"%"와 서식 문자를 사용해서 출력하기

```python
# 서식 문자
# %d(정수), %f(실수), %s(문자, 문자열)를 사용해 출력 서식을 지정할 수 있고 출력 서식 문자를 제외한 나머지 문자들은 입력된 그대로 출력된다.
# 출력 서식의 형식
# %[-][0][n][.m]서식문자 => [] 안의 내용은 생략할 수 있다.
# - : 전체 자리수(n)가 지정된 경우 데이터를 왼쪽으로 맞춰 출력한다.
# 0 : 전체 자리수(n)가 지정된 경우 숫자 데이터의 왼쪽에 남는 자리에 "0"을 채워 출력한다.
# n : 전체 자리수
# .m : 실수를 출력할 경우 지정하는 소수점 아래 자리수
```

```python
print(100)
print(1)
print(10000)
```

```python
# print("출력 서식" % 출력할 데이터)
print("%5d" % 100)
print("%5d" % 1)
print("%-5d" % 1)
print("%05d" % 1)
print("%5d" % 10000)
```

```python
print("%5s" % "abc")
print("%5s" % "a")
print("%-5s" % "a")
print("%05s" % "a") # 파이썬은 문자열 서식에서 "0"이 무시된다.
print("%5s" % "abcde")
```

```python
print("%6.2f" % 123.45)
print("%6.2f" % 123.456) # 소수점 아래 숫자는 잘리는 자리에서 반올림시켜 출력한다. => 원래 데이터는 변경되지 않는다.
print("%6.2f" % 123.4)   # 남는 자리에는 "0"이 채워져서 출력된다.
print("%4.0f" % 60.4)
print("%4.0f" % 59.5)
```

```python
# 출력 서식에 맞춰 출력할 데이터가 2건 이상이라면 출력할 데이터를 ()로 묶어야 한다.
print("%d + %d = %d" % (5, 3, 5 + 3)) # 덧셈
print("%d - %d = %d" % (5, 3, 5 - 3)) # 뺄셈
print("%d * %d = %d" % (5, 3, 5 * 3)) # 곱셈
# 파이썬은 "/" 연산자로 나눗셈을 하면 소수점 아래까지 계산된다.
print("%d / %d = %d" % (5, 3, 5 / 3)) # 나눗셈 => 나눗셈 결과를 %d를 사용해서 정수로 출력하라 했기 때문에 정수만 출력된다.
print("%d / %d = %f" % (5, 3, 5 / 3)) # 나눗셈
# 파이썬은 "//" 연산자로 나눗셈을 하면 몫만 계산된다.
print("%d // %d = %d" % (5, 3, 5 // 3)) # 나눗셈
print("%d // %d = %f" % (5, 3, 5 // 3)) # 나눗셈 => 나눗셈 결곽 계산된 몫을 실수로 출력하라 했으므로 실수로 출력된다.
# 출력 서식을 지정할 때 "%" 자체를 출력하고 싶으면 연속해서 2개를 써야한다.
print("%d %% %d = %d" % (5, 3, 5 % 3)) # 나머지
print("%d의 %d승 = %d" % (2, 10, 2 ** 10)) # 거듭제곱
print("루트 %d = %f" % (2, 2 ** 0.5))
```

```python
# 산술 연산자 우선순위 : () => ** => *, /, //, % => +, -
```

```python
# 문자열 서식을 지정하는 format() 메소드와 {}를 사용해서 출력하기
# "%"를 이용해서 서식을 지정할 때는 %d, %f, %s에 데이터가 대입되었지만 {}를 사용하면 {}에 데이터가 대입되서 출력된다.
print("%d + %d = %d" % (5, 3, 5 + 3))
print("{} + {} = {}".format(5, 3, 5 + 3))
print("{} - {} = {}".format(5, 3, 5 - 3))
print("{} * {} = {}".format(5, 3, 5 * 3))
print("{} / {} = {}".format(5, 3, 5 / 3))
print("{} // {} = {}".format(5, 3, 5 // 3))
print("{} % {} = {}".format(5, 3, 5 % 3))
print("{}의 {}승 = {}".format(2, 10, 2 ** 10))
```

```python
# format() 메소드와 출력할 데이터의 인덱스를 사용해서 출력하기 => 인덱스는 0부터 시작된다.
print("{} {}".format("홍길동", "바보"))
# 인덱스를 지정하지 않은 상태에서 아래와 같이 {}와 출력할 데이터의 개수가 다르면 에러가 발생된다. => {}가 남으면 에러가 발생된다.
# print("{} {} {}".format("홍길동", "바보")) # 에러
# {}가 모자라면 에러는 발생되지 않지만 남는 데이터가 무시된다.
print("{}".format("홍길동", "바보")) # 남는 데이터 무시
```

```python
print("{0} {1}".format("홍길동", "바보"))
print("{1} {0}".format("홍길동", "바보"))
# 인덱스가 지정되면 출력할 데이터와 {}의 개수가 달라도 인덱스의 범위만 벗어나지 않으면 정상적으로 출력된다.
print("{0} {1} {0}".format("홍길동", "바보"))
# 인덱스는 부분적으로 지정하면 에러가 발생된다.
# print("{0} {1} {}".format("홍길동", "바보")) # 에러
# 인덱스가 범위를 벗어나서 지정되면 에러가 발생된다.
# print("{0} {1} {2}".format("홍길동", "바보")) # 에러
```

```python
# 출력 서식과 format() 메소드의 출력할 데이터의 인덱스를 사용해 출력하기
# {인덱스:출력서식} => 출력 서식 앞에 "%"는 붙이지 않는다.
print("%s님은 %d살 입니다." % ("홍길동", 20))
print("{}님은 {}살 입니다.".format("홍길동", 20))
print("{0}님은 {1}살 입니다.".format("홍길동", 20))
print("{0:10s}님은 {1:03d}살 입니다.".format("홍길동", 20))
```

```python
# "<", ">", "^"를 사용해서 문자열 정렬 방법을 지정할 수 있다.
print("{0:<10s}님은 {1:03d}살 입니다.".format("홍길동", 20))
print("{0:>10s}님은 {1:03d}살 입니다.".format("홍길동", 20))
print("{0:^10s}님은 {1:03d}살 입니다.".format("홍길동", 20))
```

```python
# "<", ">", "^" 앞에 문자열 출력시 남는 공백에 채워지는 채움 문자를 지정할 수 있다.
print("{0:_<10s}님은 {1:03d}살 입니다.".format("홍길동", 20))
print("{0:+>10s}님은 {1:03d}살 입니다.".format("홍길동", 20))
print("{0:&^10s}님은 {1:03d}살 입니다.".format("홍길동", 20))
```

```python
# f 접두어를 사용한 서식 지정
name = "임꺽정"
age = 35
print(f"{name}님은 {age}살 입니다.")
```
